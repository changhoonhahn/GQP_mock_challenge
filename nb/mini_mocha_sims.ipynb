Lgal mini mocha data set looks strange because the stellar mass and redshift are highly correlated. All of the $r$ magnitudes are roughly 19.5

```python
import os
import numpy as np 
from gqp_mc import data as Data 
# --- plotting --- 
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.patches as patches
mpl.rcParams['text.usetex'] = True
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['axes.linewidth'] = 1.5
mpl.rcParams['axes.xmargin'] = 1
mpl.rcParams['xtick.labelsize'] = 'x-large'
mpl.rcParams['xtick.major.size'] = 5
mpl.rcParams['xtick.major.width'] = 1.5
mpl.rcParams['ytick.labelsize'] = 'x-large'
mpl.rcParams['ytick.major.size'] = 5
mpl.rcParams['ytick.major.width'] = 1.5
mpl.rcParams['legend.frameon'] = False
```

```python
photo, meta = Data.Photometry(sim='lgal', noise='legacy', lib='bc03', sample='mini_mocha') 
```

```python
fig = plt.figure()
sub = fig.add_subplot(111)
sub.hist(meta['redshift'], range=(0., 0.5), bins=20)
sub.set_xlabel('redshift', fontsize=20)
sub.set_xlim(0., 0.5)
```

```python
fig = plt.figure()
sub = fig.add_subplot(111)
sub.hist(meta['logM_total'], range=(9., 12.), bins=12)
sub.set_xlabel(r'$\log M_*$', fontsize=20)
sub.set_xlim(9., 12.)
```

```python
r_mag = 22.5 - 2.5 * np.log10(photo['flux_r_true']) # true r-band magnitude
```

```python
fig = plt.figure()
sub = fig.add_subplot(111)
sub.hist(r_mag, range=(19, 20.), bins=12)
sub.set_xlabel(r'$r$ magnitude', fontsize=20)
sub.set_xlim(19, 20.)
```

```python
fig = plt.figure()
sub = fig.add_subplot(111)
sub.scatter(meta['redshift'], meta['logM_total'], c='C0', s=4)
sub.set_xlabel('redshift', fontsize=20)
sub.set_xlim(0., 0.5)
sub.set_ylabel(r'$\log M_*$', fontsize=20)
sub.set_ylim(9., 12.)
```
