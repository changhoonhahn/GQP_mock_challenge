```python
import os 
import h5py
import numpy as np 
# -- gqp_mc --
from gqp_mc import util as UT
# -- plotting --
import matplotlib as mpl 
import matplotlib.pyplot as plt 
mpl.rcParams['text.usetex'] = True
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['axes.linewidth'] = 1.5
mpl.rcParams['axes.xmargin'] = 1
mpl.rcParams['xtick.labelsize'] = 'x-large'
mpl.rcParams['xtick.major.size'] = 5
mpl.rcParams['xtick.major.width'] = 1.5
mpl.rcParams['ytick.labelsize'] = 'x-large'
mpl.rcParams['ytick.major.size'] = 5
mpl.rcParams['ytick.major.width'] = 1.5
mpl.rcParams['legend.frameon'] = False
%matplotlib inline
```

```python
dir_fbgs = '/Users/ChangHoon/data/feasiBGS/'
```

```python
# read gama-legacy file 
gleg = h5py.File(os.path.join(dir_fbgs, 'GAMAdr3.g15.LEGACYdr7.v2.hdf5'), 'r') 
```

```python
# lets calculate fiber flux fraction in g, r, and z bands 
f_fiber_g = gleg['legacy-photo/fiberflux_g'][...] / gleg['legacy-photo/flux_g'][...]
f_fiber_r = gleg['legacy-photo/fiberflux_r'][...] / gleg['legacy-photo/flux_r'][...]
f_fiber_z = gleg['legacy-photo/fiberflux_z'][...] / gleg['legacy-photo/flux_z'][...]
```

```python
fig = plt.figure(figsize=(10,5))
sub = fig.add_subplot(121)
sub.scatter(f_fiber_g, f_fiber_r, c='C0', s=1)
sub.plot([0., 1.], [0., 1.], c='k', ls='--', zorder=10)
sub.set_xlabel(r'$f^g_{\rm fiber}$', fontsize=20)
sub.set_xlim(0., 1.)
sub.set_ylabel(r'$f^r_{\rm fiber}$', fontsize=20)
sub.set_ylim(0., 1.)

sub = fig.add_subplot(122)
sub.scatter(f_fiber_z, f_fiber_r, c='C0', s=1)
sub.plot([0., 1.], [0., 1.], c='k', ls='--', zorder=10)
sub.set_xlim(0., 1.)
sub.set_xlabel(r'$f^z_{\rm fiber}$', fontsize=20)
sub.set_ylim(0., 1.)
```
