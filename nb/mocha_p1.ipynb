# assign DECaLS like noise to photometry
First lets read in a small subset of DECaLS photometry

```python
bgs_targets = h5py.File(os.path.join(UT.dat_dir(), 'bgs.1400deg2.rlim21.0.hdf5'), 'r')
```

```python
bgs_fluxes, bgs_mags, bgs_ivars = [], [], []
for b in ['g', 'r', 'z', 'w1', 'w2']: 
    bgs_fluxes.append(bgs_targets['flux_%s' % b][...])
    bgs_ivars.append(bgs_targets['flux_ivar_%s' % b][...])
    
    bgs_mags.append(22.5 - 2.5 * np.log10(bgs_targets['flux_%s' % b][...]))
    
bgs_fluxes = np.array(bgs_fluxes)
bgs_ivars = np.array(bgs_ivars)
bgs_mags = np.array(bgs_mags)
```

```python
flux_ivars = []
for i in range(5): 
    med, bins, _ = scipy.stats.binned_statistic(
        np.log10(bgs_fluxes[i])[np.isfinite(np.log10(bgs_fluxes[i]))], 
        np.log10(bgs_ivars[i])[np.isfinite(np.log10(bgs_fluxes[i]))], statistic='median', bins=40, range=(-1, 3))
    ibin = np.argmin(np.abs(fluxes[:,i][:,None] - 10**(0.5 * (bins[1:] + bins[:-1]))), axis=1)
    flux_ivars.append(10**(med[ibin]))
flux_ivars = np.array(flux_ivars).T
```

```python
fig = plt.figure(figsize=(12, 8))
for i in range(5): 
    sub = fig.add_subplot(2,3,i+1)
    sub.scatter(bgs_fluxes[i][::100], bgs_ivars[i][::100], s=1)
    sub.scatter(fluxes[:,i], flux_ivars[:,i], c='C1')
    sub.set_xlim(1e-1, 100)
    sub.set_xscale('log')
```

```python
fluxes_obs = fluxes + flux_ivars**-0.5 * np.random.randn(fluxes.shape[0], fluxes.shape[1])
```

```python
fig = plt.figure(figsize=(6,6))
sub = fig.add_subplot(111)
sub.scatter(fluxes[:,0]/fluxes[:,1], fluxes[:,1]/fluxes[:,2], c='C0')
sub.scatter(fluxes_obs[:,0]/fluxes_obs[:,1], fluxes_obs[:,1]/fluxes_obs[:,2], c='C1')
sub.set_xlabel('$g-r$', fontsize=25)
sub.set_xlim(0., 0.7)
sub.set_ylabel('$r-z$', fontsize=25)
sub.set_ylim(0., 0.8)
```
